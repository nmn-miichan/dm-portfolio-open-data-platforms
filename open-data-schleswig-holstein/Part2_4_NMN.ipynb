```python
from ckanapi import RemoteCKAN
import requests
import pandas as pd
import io
import json
```

```python
ckan_url = 'https://opendata.schleswig-holstein.de/'  
ckan = RemoteCKAN(ckan_url)
```

# 1. Organization
### 1.1. Get the list of organizations

```python
organizations = ckan.action.organization_list()
```

```python
print("There are",len(organizations),"organizations in total.")
```

```python
organizations_json = json.dumps(organizations, indent=4)
print(organizations_json)
```

### 1.2. Get details of a specific organization

```python
bundesnetzagentur = ckan.action.organization_show(id='bundesnetzagentur')
bundesnetzagentur_json = json.dumps(bundesnetzagentur, indent=4, ensure_ascii=False)
print(bundesnetzagentur_json)
```

### 2.3. Search for packages

```python
# Example: search for packages related to authorities ('behörde')
search_results = ckan.action.package_search(q='behörde')

for result in search_results['results']:
    print(f"Name: {result['name']}")
    print(f"Title: {result['title']}")
    print(f"ID: {result['id']}")
    print(f"URL: {result['url']}")
    print(f"Description: {result['notes']}\n\n")
```

```python
# More information about the search results
search_results_json = json.dumps(search_results, indent=4, ensure_ascii=False)
print(search_results_json)
```

# 3. Resource
### 3.1. Get the list of resources from a specific package

```python
dataset = ckan.action.package_show(id='kfz-kennzeichen-des-landes-schleswig-holstein')

resources = dataset.get('resources', [])
```

```python
for resource in resources:
    print(f"Name: {resource['name']}")
    print(f"ID: {resource['id']}")
    print(f"URL: {resource['url']}\n")
```

### 3.2. Read data from resource (csv file)

```python
#Example from above
csv_url = dataset['resources'][0]['url']
response = requests.get(csv_url)
csv_data = response.content.decode('utf-8')
df = pd.read_csv(io.StringIO(csv_data))
pd.set_option('display.max_colwidth', None)
df
```
